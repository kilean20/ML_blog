```python
import numpy as np

from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

np.random.seed(1)
```

```python
# Enabling the `widget` backend.
# This requires jupyter-matplotlib a.k.a. ipympl.
# ipympl can be install via pip or conda.
import matplotlib.pyplot as plt
```

$$
\begin{eqnarray}
\boldsymbol{R} &=& |\boldsymbol{x}| \\
{f} &=& \frac{\sin(4\pi \boldsymbol{R})}{4\pi \boldsymbol{R}}
\end{eqnarray}
$$

```python
from mpl_toolkits.mplot3d import axes3d
%matplotlib widget
```

```python
def f(x):
    """The function to predict."""
    R = np.sqrt(x[:,0]**2 + x[:,1]**2) + 0.0001
    return np.sin(4*np.pi*R)/(4*np.pi*R)
```

```python
X =  (np.random.rand(128,2)-0.5)*2
```

```python
X.shape
```

```python
y = f(X).ravel()
```

```python
kernel = C(1.0, (1e-3, 1e3)) * RBF(1, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
```

### Before Fit

```python
gp.score(X, y)
```

```python
gp.kernel
```

### After Fit

```python
gp.fit(X, y)
```

```python
gp.score(X, y)
```

```python
gp.kernel_
```

```python
x = np.linspace(-1, 1, 512)
x_mesh = np.array(np.meshgrid(x, x))
x = np.array(x_mesh).T.reshape(-1, 2)
```

```python
y_pred, sigma = gp.predict(x, return_std=True)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
# ax.plot_wireframe(x_mesh[0], x_mesh[1], f(x).reshape(x_mesh[0].shape).T, rstride=10, cstride=10,color='b')
surf_ground = ax.plot_surface(x_mesh[0], x_mesh[1], f(x).reshape(x_mesh[0].shape).T,alpha=0.5, label='Ground Truth')
ax.scatter(X[:,0], X[:,1], f(X),c='r', label='Training Data')

ax.set_xticks([-1, -0.5,0,0.5,1]);
ax.set_yticks([-1, -0.5,0,0.5,1]);

surf_ground._facecolors2d=surf_ground._facecolors3d
surf_ground._edgecolors2d=surf_ground._edgecolors3d
ax.legend()
```

```python
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
# ax.plot_wireframe(x_mesh[0], x_mesh[1], f(x).reshape(x_mesh[0].shape).T, rstride=10, cstride=10,color='b')
surf_ground = ax.plot_surface(x_mesh[0], x_mesh[1], f(x).reshape(x_mesh[0].shape).T,alpha=0.9, label='Ground Truth')
surf_predict = ax.plot_surface(x_mesh[0], x_mesh[1], y_pred.reshape(x_mesh[0].shape).T,alpha=0.5, label='Prediction Mean')
ax.scatter(X[:,0], X[:,1], f(X),c='r', label='Training Data')

surf_ground._facecolors2d=surf_ground._facecolors3d
surf_ground._edgecolors2d=surf_ground._edgecolors3d
surf_predict._facecolors2d=surf_predict._facecolors3d
surf_predict._edgecolors2d=surf_predict._edgecolors3d
ax.legend()

ax.set_xticks([-1, -0.5,0,0.5,1]);
ax.set_yticks([-1, -0.5,0,0.5,1]);

plt.savefig('3Dplot_GP_mean.png',dpi=180)
```

```python
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
# ax.plot_wireframe(x_mesh[0], x_mesh[1], f(x).reshape(x_mesh[0].shape).T, rstride=10, cstride=10,color='b')
surf_ground = ax.plot_surface(x_mesh[0], x_mesh[1], f(x).reshape(x_mesh[0].shape).T,alpha=1.0, label='Ground Truth')
surf_up     = ax.plot_surface(x_mesh[0], x_mesh[1], (y_pred+ 1.96 * sigma).reshape(x_mesh[0].shape).T,alpha=0.3, label='95% upper limit')
surf_down   = ax.plot_surface(x_mesh[0], x_mesh[1], (y_pred- 1.96 * sigma).reshape(x_mesh[0].shape).T,alpha=0.3, label='95% lower limit')

ax.scatter(X[:,0], X[:,1], f(X),c='r', label='Training Data')


surf_ground._facecolors2d=surf_ground._facecolors3d
surf_ground._edgecolors2d=surf_ground._edgecolors3d
surf_up._facecolors2d=surf_up._facecolors3d
surf_up._edgecolors2d=surf_up._edgecolors3d
surf_down._facecolors2d=surf_down._facecolors3d
surf_down._edgecolors2d=surf_down._edgecolors3d
ax.legend()

ax.set_xticks([-1, -0.5,0,0.5,1]);
ax.set_yticks([-1, -0.5,0,0.5,1]);

plt.savefig('3Dplot_GP.png',dpi=180)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x_mesh[0][0,:],f(x).reshape(x_mesh[0].shape).T[256,:], 'k', label='Ground Truth')
ax.plot(x_mesh[0][0,:],y_pred.reshape(x_mesh[0].shape).T[256,:], 'b-', label='Prediction')
ax.fill_between(x_mesh[0][0,:], 
                (y_pred- 1.96 * sigma).reshape(x_mesh[0].shape).T[256,:],
                (y_pred+ 1.96 * sigma).reshape(x_mesh[0].shape).T[256,:], 
                label='95% confidence interval', color='b', alpha=.5, )

xOnAxis = []
for tmp in X:
    if np.sum(tmp[1:]**2) < 0.001:
        xOnAxis.append(tmp)
xOnAxis = np.array(xOnAxis)
plt.scatter(xOnAxis[:,0], f(xOnAxis), c='r', label='Observations')
```
