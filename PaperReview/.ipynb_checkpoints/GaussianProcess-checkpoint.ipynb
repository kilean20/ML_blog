# Process
A stochastic *process* is a generalization of a probability distribution to functions

# Gaussian Process (GP) <font size="2">is</font>
- a *process* that is Gaussian
- a supervised learning,
- a non-parameteric posterior distribution estimator ( compare with parametric [point estimator](PointEstimates.ipynb)),
- with assumption of multivariate gaussian distribution prior (i.e. no observed data) over label $y = f(x)$ :

$$ p(f|x) = \mathcal{N}(f \mid \mu, \ \Sigma) $$

where $\mu$ is the mean, $\Sigma$ is the covariance matrix and $x$ is set points in the domain of the label $f$

#### (formal) Definition: GP is a collection of random variables, any finite number of which have a joint Gaussian distribution

GP is completely specified by its mean function $\mu(x)$ and covariance function $k(x,x')$ which produce the covariance matrix $\Sigma$
$$
\begin{eqnarray}
\mu(x) &=& \mathbb{E}\left[f(x)\right] \\
k(x,x') &=& \mathbb{E}\left[\left(f(x)-\mu(x)\right)\left(f(x')-\mu(x')\right)\right]
\end{eqnarray}
$$


### 1. Why Gaussian?

Gaussian distribution has nice properties
- closed under linear operator:

$$ \mathcal{L} \ f \sim \mathcal{N}(\mathcal{L} \ \mu, \ \mathcal{L} \ \Sigma) $$

- *closed under Baysian conditioning*:

$$ f \mid \mathrm{observed \ data} \sim \mathcal{N}(\mu_{*} , \ \Sigma_{*}) $$

where $\mu_{*} \neq \mu$, $\Sigma_{*} \neq \Sigma$ are the updated (by conditioning) mean and covariances

### 2.1 How to choose the mean $\mu$ (of the prior)?

- common choice is $\mu=0$ which *simplifies conditioning*
- complex modeling (e.g. NN) also possible


### 2.2 How to choose the covariance $\Sigma$ (of the prior)?

the kernal trick:

$$ \Sigma_{ij} = k(x_i,x_j) $$

A common kernal is: (squared exponential kernel) = (Gaussian kernal) = (RBF kernal):

$$ k(x_i,x_j) = \sigma^2 \left(-\frac{|x_i-x_j|}{2l^2}\right) $$

where $|x_i-x_j|$ is a distance (e.g. Euclidean L2)

### 3. Baysian conditioning

Consider two data set $x$ and $\bar{x}$. Then, by definition of GP, the data is assumed to be an instance of the prior (the joint distribution)

$$
\left[\begin{array}{c}
f\\
\bar{f}
\end{array}\right]\sim\mathcal{N}\left(\left[\begin{array}{c}
\mu\\
\bar{\mu}
\end{array}\right],\left[\begin{array}{cc}
\Sigma_{ff} & \Sigma_{f\bar{f}}\\
\Sigma_{\bar{f}f} & \Sigma_{\bar{f}\bar{f}}
\end{array}\right]\right)
$$

where $f=f(x)$, $\bar{f}=f(\bar{x})$, and $\Sigma_{f\bar{f}}=k(x,\bar{x})$.

Then, the posterior is
$$
p\left(\bar{f}\ |\ f\right) = \mathcal{N}\left(\mu_{*}, \ \Sigma_{*}\right)
$$
where
$$
\begin{eqnarray}
\mu_{*}&=&\bar{\mu}+\Sigma_{\bar{f}f}\Sigma_{ff}^{-1}\left(f-\mu\right)\\
\Sigma_{*}&=&\Sigma_{\bar{f}\bar{f}}-\Sigma_{\bar{f}f}\Sigma_{ff}^{-1}\Sigma_{f\bar{f}}
\end{eqnarray}
$$



#### 3.1 Sketch of Proof

use Bayes theorem 
$$ p\left(\bar{f}\ |\ f\right) = \frac{ p\left(\bar{f},\ f\right) } {p\left(f\right)} $$
where
$$  
p\left(f\right)=\mathcal{N}\left(\mu,\Sigma_{ff}\right)
$$
to prove 
$$p\left(\bar{f}\ |\ f\right) = \mathcal{N}\left(\mu_{*}, \ \Sigma_{*}\right)$$

### 4. (optional) noise model

consider simple noise model of data
$$ y = f(x) + \epsilon $$
where $\epsilon \sim \mathcal{N}(0,\sigma^2 I)$

Then the covarance matrix modified:
- $\Sigma_{ff} \rightarrow \Sigma_{ff} + \sigma^2 I$
- $\Sigma_{\bar{f}\bar{f}} \rightarrow \Sigma_{\bar{f}\bar{f}} + \sigma^2 I$

# Example

```python
import numpy as np
import matplotlib.pyplot as plt
int = np.vectorize(int)

def kernel(x1, x2, l=1.0, sigma=1.0):
    '''
    Isotropic squared exponential kernel. Computes 
    a covariance matrix from points in x1 and x2.
        
    Args:
        x1: Array of m points (m x d).
        x2: Array of n points (n x d).

    Returns:
        Covariance matrix (m x n).
    '''
    dist = np.sum(x1**2, 1).reshape(-1, 1) + np.sum(x2**2, 1) - 2 * np.dot(x1, x2.T)
    return sigma**2 * np.exp(-0.5 / l**2 * dist)


def plot_gp(mu, cov, x, x_train=None, y_train=None, samples=[], x_true=None, y_true=None):
    x = x.ravel()
    mu = mu.ravel()
    uncertainty = 1.96 * np.sqrt(np.diag(cov))
    
    if x_true is not None:
        plt.plot(x_true, y_true, label='ground true', lw=1, c='k')
    plt.fill_between(x, mu + uncertainty, mu - uncertainty, alpha=0.2)
    plt.plot(x, mu, label='Mean')
    if x_train is not None:
        plt.plot(x_train, y_train, 'rx', label='training data')
    for i, sample in enumerate(samples):
        plt.plot(x, sample, lw=1, ls='--', label=f'sample {i+1}')

    plt.legend()
```

###  prior  <font size="2"> ( with $\mu=0$, and gaussian kernal $\sigma=1$, $l=1$ ) </font>

```python
n_sample = 3
x = np.linspace(0, 2*np.pi, 50).reshape(-1, 1)
mu = np.zeros(x.shape)
cov = kernel(x, x)

samples = np.random.multivariate_normal(mu.ravel(), cov, n_sample)


plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plot_gp(mu, cov, x, samples=samples)
plt.xlabel('x');
plt.ylabel('f(x)');
plt.title('prior');

ax = plt.subplot(1,2,2)
im = ax.matshow(cov)
ax.tick_params(axis="x", bottom=True, top=False, labelbottom=True, labeltop=False)
indexes = int(np.linspace(0,len(x)-1,3))
labels = ['0','$\pi$','2$\pi$']
ax.set_xticks(indexes)
ax.set_xticklabels(labels)
ax.set_yticks(indexes)
ax.set_yticklabels(labels)
plt.xlabel('x');
plt.ylabel('x');
plt.title('covariance matrix')
plt.colorbar(im)
```

### posterior  <font size="2"> ( $f(x)=\sin (x)$ ) </font>

```python
from numpy.linalg import inv

def posterior_predictive(x_s, x_train, y_train, l=1.0, sigma_f=1.0, sigma_y=1e-8):
    '''  
    Computes the suffifient statistics of the GP posterior predictive distribution 
    from m training data x_train and y_train and n new inputs x_s.
    
    Args:
        x_s: New input locations (n x d).
        x_train: Training locations (m x d).
        y_train: Training targets (m x 1).
        l: Kernel length parameter.
        sigma_f: Kernel vertical variation parameter.
        sigma_y: Noise parameter.
    
    Returns:
        Posterior mean vector (n x d) and covariance matrix (n x n).
    '''
    K = kernel(x_train, x_train, l, sigma_f) + sigma_y**2 * np.eye(len(x_train))
    K_s = kernel(x_train, x_s, l, sigma_f)
    K_ss = kernel(x_s, x_s, l, sigma_f) + 1e-8 * np.eye(len(x_s))
    K_inv = inv(K)
    
    # Equation (4)
    mu_s = K_s.T.dot(K_inv).dot(y_train)

    # Equation (5)
    cov_s = K_ss - K_s.T.dot(K_inv).dot(K_s)
    
    return mu_s, cov_s
```

```python
x_train = np.linspace(0, 2*np.pi, 5).reshape(-1, 1)
y_train = np.sin(x_train)

x_true = np.linspace(0, 2*np.pi, 50).reshape(-1, 1)
y_true = np.sin(x_true)

n_sample = 1

for i in range(1,6):
    mu_s, cov_s = posterior_predictive(x, x_train[:i], y_train[:i])

    samples = np.random.multivariate_normal(mu_s.ravel(), cov_s, n_sample)

    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    plot_gp(mu_s, cov_s, x, x_train=x_train[:i], y_train=y_train[:i], samples=samples,  x_true=x_true,y_true=y_true)
    plt.xlabel('x');
    plt.ylabel('f(x)');
    plt.title('posterior given '+str(i)+' evidences');
    
    ax = plt.subplot(1,2,2)
    im=ax.matshow(cov_s,interpolation='bilinear')
    ax.tick_params(axis="x", bottom=True, top=False, labelbottom=True, labeltop=False)
    indexes = int(np.linspace(0,len(x)-1,3))
    labels = ['0','$\pi$','2$\pi$']
    ax.set_xticks(indexes)
    ax.set_xticklabels(labels)
    ax.set_yticks(indexes)
    ax.set_yticklabels(labels)
    plt.xlabel('x');
    plt.ylabel('x');
    plt.title('covariance matrix')
    plt.colorbar(im)
```

### posterior with noise  <font size="2"> ( $\sigma = 0.2$ ) </font>

```python
sigma = 0.2

x_train = np.linspace(0, 2*np.pi, 5).reshape(-1, 1)
y_train = np.sin(x_train) + sigma * np.random.randn(*x_train.shape)


for i in range(1,6):
    mu_s, cov_s = posterior_predictive(x, x_train[:i], y_train[:i], sigma_y=sigma)

    samples = np.random.multivariate_normal(mu_s.ravel(), cov_s, n_sample)

    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    plot_gp(mu_s, cov_s, x, x_train=x_train[:i], y_train=y_train[:i], samples=samples,  x_true=x_true,y_true=y_true)
    plt.xlabel('x');
    plt.ylabel('f(x)');
    plt.title('posterior given '+str(i)+' noisy evidences');
    
    ax = plt.subplot(1,2,2)
    im=ax.matshow(cov_s,interpolation='bilinear')
    ax.tick_params(axis="x", bottom=True, top=False, labelbottom=True, labeltop=False)
    indexes = int(np.linspace(0,len(x)-1,3))
    labels = ['0','$\pi$','2$\pi$']
    ax.set_xticks(indexes)
    ax.set_xticklabels(labels)
    ax.set_yticks(indexes)
    ax.set_yticklabels(labels)
    plt.xlabel('x');
    plt.ylabel('x');
    plt.title('covariance matrix')
    plt.colorbar(im)
    
```
